```python
data = pd.read_csv('cat_dog.csv')
a=data["a"]
b=data["b"]


second=[]
a=list(a)
for i in a:
    first=[]
    i=i[1:-1]
    abc=i.split(", ")
    # print(i)
    # print(abc)
    for j in abc:
        j=int(j)
        # print("_",type(int(j)))
        first.append(j)
        # for k in j:
        #     print(k)
    second.append(first)
    # print(first)
    # break
# second=np.array(second)
# first=np.array(first)
# print(b.shape)
# print(second.shape)

```

```python
x=cv2.imread(f"C:/Users/Admin/Desktop/ML/images/predict3.jpg",cv2.IMREAD_GRAYSCALE)
x=x.flatten()
x=np.array(x).reshape(1, -1)
x
```

```python
from sklearn.svm import SVC 
from sklearn.model_selection import train_test_split

a_train, a_test, b_train, b_test = train_test_split(second, b,test_size=0.2,random_state=42)

model = SVC(kernel='linear')
model.fit(a_train,b_train)
print(model.predict(x))
print(model.score(a_test, b_test))
```

```python
model = SVC(kernel='poly')
model.fit(a_train,b_train)
print(model.predict(x))
print(model.score(a_test, b_test))
```

```python
model = SVC(kernel='sigmoid')
model.fit(a_train,b_train)
print(model.predict(x))
print(model.score(a_test, b_test))
```
