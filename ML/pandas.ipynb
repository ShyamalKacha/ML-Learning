```python
import pandas as pd
import random
import string

sr_no=[]
for i in range(0,200):
    sr_no.append(i)

id=[]
name=[]
amount=[]
for i in range(0,200):
    id.append(random.randint(100,1000))
    name.append("".join(random.choices(string.ascii_lowercase, k=4)))
    amount.append(random.randint(1,10000))


data={"sr_no":sr_no,"ID":id,"Name":name,"amount":amount}
df = pd.DataFrame(data)
df
```
