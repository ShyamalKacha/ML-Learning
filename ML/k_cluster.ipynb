```python
import pandas as pd
import random
import numpy as np
a=[]
b=[]
c=[]
for i in range(0,400):
    x=round(random.uniform(3.0,10.0),2)
    a.append(x)
    b.append(x*25)
    c.append(1)
for i in range(0,400):
    x=round(random.uniform(3.0,10.0),2)
    a.append(x)
    b.append(x*15)
    c.append(0)

data={"hight":a,"weight":b,"c":c}
df = pd.DataFrame(data)
df.to_csv("k_means.csv")

```
